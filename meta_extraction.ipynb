```python
# Join the two DataFrames on the Subject column
merged_df = pd.merge(samm_annotations, samm_df[['Subject', 'Ethnic_Concised']], on='Subject', how='left')
```

```python
def categorize_emotion(emotion):
    if emotion == 'Happiness':
        return 'Positive'
    elif emotion == 'Surprise':
        return 'Surprise'
    elif emotion == 'Other':
        return 'Other'
    else:
        return 'Negative'

```

```python
# Join the two DataFrames on the Subject column
ethnic_labelled_df['RaceConcised'] = ethnic_labelled_df['Race'].apply(lambda x: 'Asian' if x >= 2 else 'Non-Asian')
ethnic_labelled_df['Subject'] = ethnic_labelled_df['Subject'].astype(str).str.zfill(3)
merged_df = pd.merge(samm_annotations, ethnic_labelled_df[['Subject', 'Age', 'Gender', 'Race', 'RaceConcised']], on='Subject', how='left')
merged_df['EstimatedEmotionConcised'] = merged_df['Estimated Emotion'].apply(categorize_emotion)
```
